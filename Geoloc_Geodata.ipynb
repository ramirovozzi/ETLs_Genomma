```python
import pandas as pd
import os
from datetime import date
import numpy as np

año = '2022'
mes = '02'

direc_noe = 'C:/Users/rvozzi/genommalabinternacional/Noelia Soledad Riccelli - Plan BI/'
direc = direc_noe + 'archivos mes/' + año + mes + '/Mercados'
os.listdir(direc)
```

```python
cols = [0,1,3,4]
col_names = ['cpa', 'desc','localidad','provincia']
cpa =    pd.read_csv(direc + '/CPA_Desc.txt', header = None, encoding='latin-1', usecols = cols, names = col_names)
listado_cpas = pd.read_csv(direc_noe + 'archivos mes/' + año + mes + '/CPAs.csv', names = ['num'])
cpa = pd.merge(cpa,
               listado_cpas,
               left_on = 'cpa',
               right_on = 'num',
               how = 'inner')
```

```python
#CAMBIO EL NOMBRE DE LAS PROVINCIAS QUE ESTAN MAL ESCRITAS
cpa.loc[cpa['provincia']=='LA PLATA', 'provincia'] = 'BUENOS AIRES'
cpa.loc[cpa['provincia']=='SUBURBANO', 'provincia'] = 'BUENOS AIRES'
cpa.loc[cpa['provincia']=='CAPITAL FEDERA', 'provincia'] = 'CAPITAL FEDERAL'
cpa.loc[cpa['provincia']=='TIERRA DEL FUE', 'provincia'] = 'TIERRA DEL FUEGO'
cpa.loc[cpa['provincia']=='SANTIAGO DEL E', 'provincia'] = 'SANTIAGO DEL ESTERO'
pd.unique(cpa['provincia'])
```

```python
# TODO limpiar la columna localidad usando el '(' como split
cpa['localidad'] = cpa['localidad'].str.split('(', expand=True)[0]
cpa.loc[cpa['localidad']=='RESISTENCIA SUC.N* 2/ UNID. PO ', 'localidad'] = 'RESISTENCIA'
```

```python
cpa['desc'] = cpa['desc'].str.split('-', expand=True)[0]
```
